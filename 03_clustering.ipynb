```python
import os
import sys
import warnings
from pathlib import Path

# Configure environment
warnings.filterwarnings('ignore')
current_dir = Path.cwd()
src_dir = current_dir / "src"
if src_dir.exists():
    sys.path.insert(0, str(src_dir))

# Configure quick analysis (without grid search)
from clustering.config import RANDOM_FOREST_CONFIG
RANDOM_FOREST_CONFIG['enabled'] = False

# Run analysis
from clustering import main_clustering_workflow
print("Running clustering analysis...")
results_df, workflow_info = main_clustering_workflow()

# Display results
print("\nRESULTS:")
print(results_df.round(4))

best_idx = results_df['Sil_test'].idxmax()
best = results_df.loc[best_idx]
print(f"\nBEST: {best['Modalidad']} | Clusters: {best['k_opt']} | Silhouette: {best['Sil_test']:.4f}")

# Basic visualization
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(range(len(results_df)), results_df['Sil_test'])
plt.title('Silhouette Score per Modality')
plt.xticks(range(len(results_df)), results_df['Modalidad'].str.replace('_', ' '), rotation=45)
plt.subplot(1, 2, 2)
plt.bar(range(len(results_df)), results_df['DB_test'], color='orange')
plt.title('Davies-Bouldin Index per Modality')
plt.xticks(range(len(results_df)), results_df['Modalidad'].str.replace('_', ' '), rotation=45)
plt.tight_layout()
plt.show()

print("Analysis completed!")
```
